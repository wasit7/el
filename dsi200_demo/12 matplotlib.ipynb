```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#plt.style.use('ggplot')
plt.style.use('seaborn-paper')
```

# Line Chart

```python
x=np.arange(10)
y=x*x
plt.plot(x,y);
```

# Color Marker and Stroke
[line style or marker](https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html)

```python
x=np.linspace(start=-3.14159,stop=3.14159,num=10)
y=np.sin(x)
plt.plot(x,y,'--ro');
```

# Title Axes and Labels

```python
x=np.linspace(start=-3.14159,stop=3.14159,num=20)
y=np.sin(x)
plt.title('My title of a line graph $V=sin(t)$')
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.plot(x,y,'--');
```

# Multiple-Line

```python
x=np.linspace(start=-3.14159,stop=3.14159,num=20)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'--ro',x,z,'-gs');
```

# Area Chart

```python
x=np.linspace(start=0,stop=6,num=20)
y=x
z=y+x
plt.fill_between(x, y);
plt.fill_between(x, z, y);
```

# Bar Chart
[color](https://matplotlib.org/3.3.1/tutorials/colors/colors.html)

```python
N=5
height=[80,90,100,120,130]
std=[8,5,2,10,15]
index=np.arange(N)
plt.bar(x=index, height=height, yerr=std, color='#f3cada');
```

# Stacked Bar Chart

```python
N=5
boy_height = (80,90,95,100,110)
girl_height = (80,90,100,120,130)
index=np.arange(N)
plt.bar(x=index, height=boy_height, color='#0fa0b0');
plt.bar(x=index, height=girl_height, color='#f3cada',bottom=boy_height);
plt.xticks(ticks=index,labels=['one','two','three','four','five']);
plt.xlabel('Age (year)')
plt.ylabel('Height (cm)')
```

# rcParams
[customizing rcParams](https://matplotlib.org/tutorials/introductory/customizing.html)

```python
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['figure.figsize'] = 10,6

x=np.linspace(start=-3.14159,stop=3.14159,num=20)
y=np.sin(x)
plt.title('My title of a line graph $V=sin(t)$')
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.plot(x,y,'--');
plt.show()
```

# Scatter Plot

```python
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['figure.figsize'] = 12,6
# Fixing random state for reproducibility
np.random.seed(2)


N =100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.show()
```
