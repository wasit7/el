```python
import pandas as pd
```

```python
df=pd.read_csv('BreadBasket_DMS.csv')
df.head()
```

```python
df['item_id']=df['Item'].astype('category').cat.codes
df.drop(['Date','Time'],axis=1, inplace=True)
df3=df.drop_duplicates(subset=['item_id'])\
    .drop(['Transaction'],axis=1)\
    .set_index('item_id')\
    .sort_index()
```

```python
df3
```

```python
df3.to_csv('sku.csv')
```

```python
df2=df.drop_duplicates(subset=['Transaction'])
df2.set_index('Transaction',inplace=True)
df2.index.names = ['id']
df2.drop('Item',axis=1,inplace=True)
df2
```

```python
df2.to_csv('transaction.csv')
```
